```python
from math import log
import csv

f = open(r'D:\pythonclass\git_project\Book1.csv','r')
data = list(csv.reader(f))
import numpy as np
data2 = np.array(data[1:],dtype=np.float)
rng=[i for i in range(len(data2))]       #rng=[0 to 69]
listt=[i for i in range(len(data[0])-1)] #listt=[0,1,2,3]
print(np.array(data2[:,0].mean()))
leaf1=[]
    
    
def counting(k,l):
    count30=0
    count31=0
    count32=0
    count40=0
    count41=0
    count42=0
    for j in k:
        if j[-1]==0:
            count30 +=1
        elif j[-1]==1:
            count31 +=1
        elif j[-1]==2:
            count32 +=1
    count3=[count30,count31,count32]
    for j in l:
        if j[-1]==0:
            count40 +=1
        elif j[-1]==1:
            count41 +=1
        elif j[-1]==2:
            count42 +=1  
    count4=[count40,count41,count42]
    return (count3,count4)

def find_entropy(l,g):
    b=np.array(g[:,l].mean())
    data3=np.array([d for d in g if d[l]<b])
    data4=np.array([d for d in g if d[l]>=b])
   
    count3,count4=counting(data3,data4)
    entro3=0
    entro4=0
    for x in count3:
        y=x/sum(count3)
        if y==0:
            entro3 +=1
        else:
            entro3 += -y*log(y,2) 
    for x in count4:
        y=x/sum(count4)
        if y==0:
            entro4 +=1
        else:
            entro4 += -y*log(y,2)  
    entro=entro3+entro4
    return entro

datax=[]    

def nodee(lista,dataa):
    entro_list=[]
    for i in listt:
        entro_list.append(find_entropy(i,dataa))
    print(entro_list)
    for i in lista:
        if find_entropy(i,dataa)==min(entro_list):
            q=i
    lista.pop(q)
    list1=lista
    print(q)
    mean1=dataa[:,q].mean()
    print(mean1)
    datax.append(np.array([d for d in dataa if d[q]<mean1]))
    data5=datax[-1]
    datax.append(np.array([d for d in dataa if d[q]>=mean1]))
    data6=datax[-1]
    count5,count6=counting(data5,data6)
    
    if count5[0]==len(data5):
        leaf1.append([1,"low"])
    elif count5[1]==len(data5):
        leaf1.append([1,"medium"])
    elif count5[2]==len(data5):
        leaf1.append([1,"high"])
    else:
        nodee(list1,data5)
    
    lista.pop(q)
    list2=lista
    if count6[0]==len(data6):
        leaf2=[1,"low"]
    elif count6[1]==len(data6):
        leaf2=[1,"medium"]
    elif count6[2]==len(data6):
        leaf2=[1,"high"]
    else:
        nodee(list2,data6)
nodee(listt,data2)
```
